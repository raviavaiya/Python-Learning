Joining Two DataFrame

```python
data={
    'Roll':[1,2,3,4],
    'Name':['a','b','c','d']
    
}
df1=pd.DataFrame(data)
df1
```

```python
data={
    'Roll':[1,2,4],
    'Mark':[100,200,300]
}
df2=pd.DataFrame(data)
df2
```

```python
pd.merge(df1,df2,on='Roll',how='inner')
```

```python
df1.merge(df2,on='Roll',how='inner')
```

```python
df1.merge(df2,on='Roll',how='right')
```

```python
df1.merge(df2,on='Roll',how='left')
```
