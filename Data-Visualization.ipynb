```python
import matplotlib.pyplot as  plt
import pandas as pd
```

```python
Movies_df=pd.read_csv("Marvel Movies.csv")
```

```python
year=[2021,2022,2023,2024,2025]
price=[12,52,3,5,1]
```

```python
plt.plot(year,price,marker='o',color='g',label='Line Plot')
for i in range(len(year)):
    plt.text(year[i], price[i], f'{price[i]}', fontsize=9, ha='right', va='bottom')
plt.title('Price of Gold')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.bar(year,price,color='blue',label='Line Plot')
plt.title('Price of Gold')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.savefig('bar.png')
plt.show()
```

```python
plt.scatter(year,price,color='g',label='Line Plot')
for i in range(len(year)):
    plt.text(year[i], price[i], f'{price[i]}', fontsize=9, ha='right', va='bottom')
plt.title('Price of Gold')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.savefig('sacter.png')
plt.show()
```
